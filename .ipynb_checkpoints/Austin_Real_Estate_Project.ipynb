```python
import pandas as pd
import numpy as np

# Load dataset
#From Sheets link
url = "https://docs.google.com/spreadsheets/d/13Wzh4NoblTQIJSLwg-13z_QcnABLG9324SToI24Chtw/gviz/tq?tqx=out:csv&sheet=Sheet1"
df = pd.read_csv(url)

print(df.head())

#Manual local data upload
#file_path = 'austinHousingData.csv'
#df = pd.read_csv(file_path)


# Display shape and first few rows
df_shape = df.shape
df_head = df.head()
df_shape, df_head
```

```python
# Classify columns into categorical, discrete, continuous
col_types = {}
for col in df.columns:
    dtype = df[col].dtype
    if dtype == 'object':
        col_types[col] = 'categorical'
    elif np.issubdtype(dtype, np.integer):
        col_types[col] = 'discrete'
    elif np.issubdtype(dtype, np.floating):
        col_types[col] = 'continuous'
    elif dtype == 'bool':
        col_types[col] = 'categorical'
    else:
        col_types[col] = 'unknown'

col_types
```

```python
# Stats for categorical columns
cat_stats = {}
for col, t in col_types.items():
    if t == 'categorical':
        unique_vals = df[col].dropna().unique()
        value_counts = df[col].value_counts(dropna=True)
        mode = value_counts.idxmax() if not value_counts.empty else None
        cat_stats[col] = {
            'num_unique': len(unique_vals),
            'mode': mode,
            'value_counts_top5': value_counts.head(5)
        }
cat_stats
```

```python
# Stats for numerical columns
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
stats = []
for col in numeric_cols:
    series = df[col].dropna().astype(float)
    stats.append({
        'Column': col,
        'Count': len(series),
        'Min': series.min(),
        'Max': series.max(),
        'Range': series.max() - series.min(),
        'Mean': series.mean(),
        'Median': series.median(),
        'StdDev': series.std()
    })

import pandas as pd
numeric_stats_df = pd.DataFrame(stats)
numeric_stats_df
```

```python
# Check missing values
missing_counts = df.isnull().sum()
missing_counts[missing_counts > 0]
```

```python
# Replace unrealistic zero values for bedrooms and bathrooms with median
bed_median = df[df['numOfBedrooms']>0]['numOfBedrooms'].median()
bath_median = df[df['numOfBathrooms']>0]['numOfBathrooms'].median()

# Perform replacement
df['numOfBedrooms'] = df['numOfBedrooms'].replace(0, bed_median)
df['numOfBathrooms'] = df['numOfBathrooms'].replace(0, bath_median)

(bed_median, bath_median)
```

```python
# Compute correlation matrix for numeric variables
df_numeric = df.select_dtypes(include=['int64','float64'])
corr_matrix = df_numeric.corr(method='pearson')

# Identify pairs with high correlation
high_corrs = []
for i, col1 in enumerate(corr_matrix.columns):
    for j, col2 in enumerate(corr_matrix.columns):
        if j <= i: continue
        coef = corr_matrix.loc[col1, col2]
        if abs(coef) > 0.7:
            high_corrs.append((col1, col2, coef))

# Sort by absolute value
high_corrs_sorted = sorted(high_corrs, key=lambda x: -abs(x[2]))
high_corrs_sorted
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np

# Evaluate regression fits for high correlations
results = {}
pairs = [(c1, c2) for c1, c2, corr in high_corrs_sorted]
for x_col, y_col in pairs:
    X = df[[x_col]].values
    y = df[y_col].values
    r2_scores = {}
    for deg in [1,2,3]:
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        X_poly = poly.fit_transform(X)
        model = LinearRegression()
        model.fit(X_poly, y)
        y_pred = model.predict(X_poly)
        r2_scores[f'degree_{deg}'] = r2_score(y, y_pred)
    results[(x_col, y_col)] = r2_scores

results
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix for numeric variables
df_numeric = df.select_dtypes(include=['int64','float64'])
corr_matrix = df_numeric.corr(method='pearson')

print("Pearson correlation matrix for quantitative data:")
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Pearson Correlation Matrix')
plt.show()

# Identify pairs with high correlation
high_corrs = []
for i, col1 in enumerate(corr_matrix.columns):
    for j, col2 in enumerate(corr_matrix.columns):
        if j <= i: continue
        coef = corr_matrix.loc[col1, col2]
        if abs(coef) > 0.7:
            high_corrs.append((col1, col2, coef))

# Sort by absolute value
high_corrs_sorted = sorted(high_corrs, key=lambda x: -abs(x[2]))
print("\nHighly correlated pairs (correlation coefficient > 0.7):")
display(high_corrs_sorted)

# Evaluate regression fits for high correlations
results = {}
pairs = [(c1, c2) for c1, c2, corr in high_corrs_sorted]
for x_col, y_col in pairs:
    X = df[[x_col]].values
    y = df[y_col].values
    r2_scores = {}
    for deg in [1,2,3]:
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        X_poly = poly.fit_transform(X)
        model = LinearRegression()
        model.fit(X_poly, y)
        y_pred = model.predict(X_poly)
        r2_scores[f'degree_{deg}'] = r2_score(y, y_pred)
    results[(x_col, y_col)] = r2_scores

print("\nR-squared scores for different polynomial regression fits:")
display(results)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Assuming high_corrs_sorted and df are available from previous cells

pairs_to_plot = [(c1, c2) for c1, c2, corr in high_corrs_sorted]

for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_col, y=y_col, data=df, alpha=0.6)

    X = df[[x_col]].values
    y = df[y_col].values

    # Sort data for plotting smooth lines
    sort_idx = np.argsort(X.flatten())
    X_sorted = X[sort_idx]
    y_sorted = y[sort_idx]

    # Plot Linear, Quadratic, and Cubic fits
    colors = ['red', 'green', 'purple']
    degrees = [1, 2, 3]
    for deg, color in zip(degrees, colors):
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        X_poly = poly.fit_transform(X)
        model = LinearRegression()
        model.fit(X_poly, y)
        y_pred = model.predict(poly.fit_transform(X_sorted))
        plt.plot(X_sorted, y_pred, color=color, label=f'Degree {deg} Fit')

    plt.title(f'{y_col} vs {x_col} with Polynomial Fits')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.show()
```
